```python
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Dropout, Flatten, Dense, BatchNormalization
from tensorflow.keras.models import Sequential

train_datagen = ImageDataGenerator(
    rescale=1./255,
    shear_range=0.2,
    zoom_range=0.2,
    horizontal_flip=True
)
test_datagen = ImageDataGenerator(rescale=1./255)

train_set = train_datagen.flow_from_directory(
    '/kaggle/input/orange-diseases-dataset/train',
    target_size=(128, 128),
    batch_size=32,
    class_mode='categorical'
)
test_set = test_datagen.flow_from_directory(
    '/kaggle/input/orange-diseases-dataset/test',
    target_size=(128, 128),
    batch_size=32,
    class_mode='categorical'
)

# بناء النموذج
cnn = Sequential()
cnn.add(Conv2D(filters=32, kernel_size=3, padding='same', activation='relu', input_shape=[128, 128, 3]))
cnn.add(MaxPooling2D(pool_size=2, strides=2, padding='valid'))
cnn.add(Dropout(0.4))
cnn.add(Conv2D(filters=64, kernel_size=3, padding='same', activation='relu'))
cnn.add(MaxPooling2D(pool_size=2, strides=2, padding='valid'))
cnn.add(Dropout(0.2))
cnn.add(BatchNormalization())
cnn.add(Flatten())
cnn.add(Dense(units=128, activation='relu'))
cnn.add(Dense(units=64, activation='relu'))
cnn.add(Dropout(0.3))
cnn.add(Dense(units=32, activation='relu'))
cnn.add(Dense(units=4, activation='softmax'))  # تغيير إلى 4 وحدات و softmax

# تجميع النموذج
cnn.compile(optimizer='sgd', loss='categorical_crossentropy', metrics=['accuracy'])

# تدريب النموذج
cnn.fit(
    train_set,
    epochs=10,
    validation_data=test_set
)

# تقييم النموذج
cnn.evaluate(test_set)
```

```python
import keras
from keras.models import Model
from keras.layers import Dense, Conv2D, MaxPool2D, Flatten, Concatenate, Input

input_layer = Input(shape=(64, 64, 3))

x1 = Conv2D(32, 1, activation='relu', padding='same')(input_layer)
x2 = Conv2D(32, 3, activation='relu', padding='same')(input_layer)
x3 = Conv2D(32, 5, activation='relu', padding='same')(input_layer)

x1_pool = MaxPool2D(2, strides=2, padding='same')(x1)
x2_pool = MaxPool2D(2, strides=2, padding='same')(x2)
x3_pool = MaxPool2D(2, strides=2, padding='same')(x3)

x5 = Concatenate()([x1_pool, x2_pool, x3_pool])

x5 = Conv2D(32, 3, activation='relu', padding='same')(x5)
x5 = MaxPool2D(2, strides=2, padding='same')(x5)

x6 = Flatten()(x5)
x6 = Dense(16, activation='relu')(x6)
x6 = Dense(16, activation='relu')(x6)
x6 = Dense(16, activation='relu')(x6)
x6 = Dense(16, activation='relu')(x6)
output_layer = Dense(4, activation='softmax')(x6)


model = Model(inputs=input_layer, outputs=output_layer)
model.summary()
```
