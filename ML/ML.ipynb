```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
import time
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix, mean_squared_error, r2_score, f1_score
import numpy as np
from sklearn.ensemble import BaggingClassifier ,ExtraTreesClassifier,BaggingRegressor ,RandomForestClassifier

```

```python

# Load the dataset
df = pd.read_csv("healthcare_dataset.csv")
df.head()
```

```python
df = df.drop(columns=['Name', 'Date of Admission', 'Discharge Date'])
dtype=df.dtypes
df.head()
```

```python
df
```

```python
# Encode categorical features
for i in range(df.shape[1]):
    if df.dtypes[df.columns[i]]=="O":
        label_encoder =LabelEncoder()
        df[df.columns[i]]=label_encoder.fit_transform(df[df.columns[i]])
df
```

```python
target_column = 'Admission Type'
y = df[target_column]
y = LabelEncoder().fit_transform(y)
y
```

```python
X_train, X_test, y_train, y_test = train_test_split(df.drop(columns=[target_column]), y, test_size=0.2, random_state=42)
```

```python
from sklearn.ensemble import AdaBoostClassifier
ada_model = AdaBoostClassifier(n_estimators=50, random_state=42)
start_time = time.time()
ada_model.fit(X_train, y_train)
training_time = time.time() - start_time

y_pred = ada_model.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)
report = classification_report(y_test, y_pred)
conf_matrix = confusion_matrix(y_test, y_pred)

print("Model: AdaBoostClassifier")
print(f"Training Time: {training_time:.2f} seconds")
print(f"Accuracy: {accuracy * 100:.2f}%")
print("Classification Report:")
print(report)
print("Confusion Matrix:")
print(conf_matrix)
```

```python
# Section 2: AdaBoostRegressor
target_column_regression = 'Billing Amount'
y_reg = df[target_column_regression]

X_train_reg, X_test_reg, y_train_reg, y_test_reg = train_test_split(df.drop(columns=[target_column_regression]), y_reg, test_size=0.2, random_state=42)

from sklearn.ensemble import AdaBoostRegressor
ada_reg_model = AdaBoostRegressor(n_estimators=50, random_state=42)
start_time = time.time()
ada_reg_model.fit(X_train_reg, y_train_reg)
training_time_reg = time.time() - start_time

y_pred_reg = ada_reg_model.predict(X_test_reg)

from sklearn.metrics import mean_squared_error, r2_score
tolerance = 0.1
accuracy_within_tolerance = sum(abs(y_pred_reg - y_test_reg) <= (tolerance * y_test_reg)) / len(y_test_reg)

mse = mean_squared_error(y_test_reg, y_pred_reg)
r2 = r2_score(y_test_reg, y_pred_reg)

print("\nModel: AdaBoostRegressor")
print(f"Training Time: {training_time_reg:.2f} seconds")
print(f"Mean Squared Error: {mse}")
print(f"R^2 Score: {r2}")
print(f"Accuracy {tolerance * 100}%: {accuracy_within_tolerance * 100:.2f}%")
```

```python

# Section 3: Decision Tree Classifier
# ------------------------------------
from sklearn.tree import DecisionTreeClassifier

tree_model = DecisionTreeClassifier()
start_time = time.time()
tree_model.fit(X_train, y_train)
y_pred = tree_model.predict(X_test)

accuracy = accuracy_score(y_test, y_pred)
report = classification_report(y_test, y_pred)
conf_matrix = confusion_matrix(y_test, y_pred)

print("\nModel: Decision Tree Classifier")
print(f"Training Time: {time.time() - start_time:.2f} seconds")
print(f"Accuracy: {accuracy * 100:.2f}%")
print("Classification Report:")
print(report)
print("Confusion Matrix:")
print(conf_matrix)
```

```python

# Section 4: Support Vector Classifier (SVC)
# -------------------------------------------
from sklearn.svm import SVC

svc_model = SVC()
start_time = time.time()
svc_model.fit(X_train, y_train)
y_pred = svc_model.predict(X_test)

accuracy = accuracy_score(y_test, y_pred)
report = classification_report(y_test, y_pred)
conf_matrix = confusion_matrix(y_test, y_pred)

print("\nModel: Support Vector Classifier (SVC)")
print(f"Training Time: {time.time() - start_time:.2f} seconds")
print(f"Accuracy: {accuracy * 100:.2f}%")
print("Classification Report:")
print(report)
print("Confusion Matrix:")
print(conf_matrix)
```

```python
# Section 5: Linear Regression
# -----------------------------
import time
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score

model = LinearRegression()

start_time = time.time()
model.fit(X_train, y_train)
elapsed_time = time.time() - start_time

y_pred = model.predict(X_test)

mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print("\nModel: Linear Regression")
print(f"Mean Squared Error: {mse:.2f}")
print(f"Accuracy: {accuracy:.2f}%")
print(f"Time Taken: {elapsed_time:.2f} seconds")
```

```python

# Section 6: KNeighborsClassifier
# --------------------------------
import time
import numpy as np
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score, f1_score, confusion_matrix

model = KNeighborsClassifier()

start_time = time.time()
model.fit(X_train, y_train)
elapsed_time = time.time() - start_time

y_pred = model.predict(X_test)

accuracy = accuracy_score(y_test, y_pred) * 100
f1 = f1_score(y_test, y_pred, average='weighted')
conf_matrix = confusion_matrix(y_test, y_pred)
worst_case = np.min(np.diag(conf_matrix))

print("\nModel: KNeighborsClassifier")
print(f"Accuracy: {accuracy:.2f}%")
print(f"F1 Score: {f1:.2f}")
print(f"Time Taken: {elapsed_time:.2f} seconds")
print(f"Worst Case: {worst_case}")
```

```python
# Section 7: Gaussian Naive Bayes
# --------------------------------
import time
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score, classification_report

nb_classifier = GaussianNB()

start_time = time.time()
nb_classifier.fit(X_train, y_train)
training_time = time.time() - start_time

y_pred_nb = nb_classifier.predict(X_test)

accuracy_nb = accuracy_score(y_test, y_pred_nb) * 100
report_nb = classification_report(y_test, y_pred_nb)

print("\nModel: Gaussian Naive Bayes")
print(f"Training Time: {training_time:.2f} seconds")
print(f"Accuracy: {accuracy_nb:.2f}%")
print(f"Classification Report:\n{report_nb}")
```

```python
# # Section 8: Polynomial Logistic Regression
# # -----------------------------------------
import time
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LogisticRegression
from sklearn.pipeline import Pipeline
from sklearn.metrics import accuracy_score, classification_report

model = Pipeline([('poly', PolynomialFeatures(degree=2)), ('logistic', LogisticRegression(max_iter=1000))])

start_time = time.time()
model.fit(X_train, y_train)
training_time = time.time() - start_time

y_pred_poly = model.predict(X_test)

accuracy_poly = accuracy_score(y_test, y_pred_poly) * 100
report_poly = classification_report(y_test, y_pred_poly)

print("\nModel: Polynomial Logistic Regression")
print(f"Training Time: {training_time:.2f} seconds")
print(f"Accuracy: {accuracy_poly:.2f}%")
print(f"Classification Report:\n{report_poly}")
```

```python
# # Section 9: StackingClassifier
# # -----------------------------
import time
from sklearn.ensemble import StackingClassifier, RandomForestClassifier, GradientBoostingClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report

base_models = [('rf', RandomForestClassifier(random_state=42)), ('gb', GradientBoostingClassifier(random_state=42))]
stacking_clf = StackingClassifier(estimators=base_models, final_estimator=LogisticRegression(), cv=5)

start_time = time.time()
stacking_clf.fit(X_train, y_train)
training_time = time.time() - start_time

y_pred_stack = stacking_clf.predict(X_test)

accuracy_stack = accuracy_score(y_test, y_pred_stack) * 100
report_stack = classification_report(y_test, y_pred_stack)

print("\nModel: StackingClassifier")
print(f"Training Time: {training_time:.2f} seconds")
print(f"Accuracy: {accuracy_stack:.2f}%")
print(f"Classification Report:\n{report_stack}")
```

```python

# Section 10: BaggingClassifier
# ----------------------------
bagging_model = BaggingClassifier(n_estimators=50, random_state=42)
start_time = time.time()
bagging_model.fit(X_train, y_train)
training_time = time.time() - start_time

y_pred = bagging_model.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)
report = classification_report(y_test, y_pred)
conf_matrix = confusion_matrix(y_test, y_pred)

print("\nModel: BaggingClassifier")
print(f"Training Time: {training_time:.2f} seconds")
print(f"Accuracy: {accuracy * 100:.2f}%")
print("Classification Report:")
print(report)
print("Confusion Matrix:")
print(conf_matrix)
```

```python

# Section 11: BaggingRegressor
# ----------------------------
from sklearn.model_selection import train_test_split
from sklearn.ensemble import BaggingRegressor
from sklearn.metrics import mean_squared_error, r2_score
import time

target_column_regression = 'Billing Amount'
y_reg = df[target_column_regression]
X_train_reg, X_test_reg, y_train_reg, y_test_reg = train_test_split(df.drop(columns=[target_column_regression]), y_reg, test_size=0.2, random_state=42)

bagging_reg_model = BaggingRegressor(n_estimators=50, random_state=42)
start_time = time.time()
bagging_reg_model.fit(X_train_reg, y_train_reg)
training_time_reg = time.time() - start_time

y_pred_reg = bagging_reg_model.predict(X_test_reg)

tolerance = 0.1
accuracy_within_tolerance = sum(abs(y_pred_reg - y_test_reg) <= (tolerance * y_test_reg)) / len(y_test_reg)

mse = mean_squared_error(y_test_reg, y_pred_reg)
r2 = r2_score(y_test_reg, y_pred_reg)

print("\nModel: BaggingRegressor")
print(f"Training Time: {training_time_reg:.2f} seconds")
print(f"Mean Squared Error: {mse}")
print(f"R^2 Score: {r2}")
print(f"Accuracy within {tolerance * 100}% tolerance: {accuracy_within_tolerance * 100:.2f}%")

```

```python

# Section 13: RandomForestClassifier
# ---------------------------------
import time
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report

features = df.drop(columns=['Medical Condition', 'Doctor', 'Hospital', 'Medication'])
target = df['Medical Condition']
X_train, X_test, y_train, y_test = train_test_split(features, target, test_size=0.2, random_state=42)

rf_classifier = RandomForestClassifier(random_state=42)

start_time = time.time()
rf_classifier.fit(X_train, y_train)
training_time = time.time() - start_time

y_pred = rf_classifier.predict(X_test)

accuracy = accuracy_score(y_test, y_pred)
report = classification_report(y_test, y_pred)

print("\nModel: RandomForestClassifier")
print(f"Training Time: {training_time:.2f} seconds")
print(f"Accuracy: {accuracy * 100:.2f}%")
print(f"Classification Report:\n{report}")

```
